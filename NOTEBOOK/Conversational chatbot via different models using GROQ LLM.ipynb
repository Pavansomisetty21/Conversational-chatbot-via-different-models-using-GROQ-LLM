```python
api_key="GROQ_API_KEY"
import os
from langchain.chains import LLMChain
from langchain_core.prompts import (
    ChatPromptTemplate,
    HumanMessagePromptTemplate,
    MessagesPlaceholder,
)
from langchain_core.messages import SystemMessage
from langchain.chains.conversation.memory import ConversationBufferWindowMemory
from langchain_groq import ChatGroq

def main():
    """
    This function is the main entry point of the application. 
    It sets up the Groq client, handles user interaction, and maintains chat history.
    """
    
    # Get Groq API key from environment variable
    groq_api_key = api_key
    if not groq_api_key:
        print("Error: GROQ_API_KEY is not set in environment variables.")
        return

    # Set up chat parameters
    system_prompt = input("Enter system prompt: ") or "You are a helpful AI assistant."
    model = input("Choose a model (llama-3.3-70b-versatile, mixtral-8x7b-32768,gemma2-9b-it): ") or "llama3-8b-8192"
    conversational_memory_length = int(input("Enter conversational memory length (1-10): ") or 5)

    # Initialize memory buffer
    memory = ConversationBufferWindowMemory(k=conversational_memory_length, memory_key="chat_history", return_messages=True)
    
    # Initialize Groq LangChain chat object
    groq_chat = ChatGroq(groq_api_key=groq_api_key, model_name=model)

    # Construct a chat prompt template using various components
    prompt = ChatPromptTemplate.from_messages(
        [
            SystemMessage(content=system_prompt),  # Persistent system prompt

            MessagesPlaceholder(variable_name="chat_history"),  # Chat history placeholder

            HumanMessagePromptTemplate.from_template("{human_input}")  # User input template
        ]
    )

    # Create a conversation chain
    conversation = LLMChain(
        llm=groq_chat,
        prompt=prompt,
        verbose=True,
        memory=memory
    )

    print("\nChatbot is ready! Type 'exit' to stop the conversation.")

    # Start the chat loop
    while True:
        user_input = input("\nYou: ")
        if user_input.lower() == "exit":
            print("Goodbye!")
            break

        # Get chatbot response
        response = conversation.predict(human_input=user_input)

        # Store conversation history
        memory.save_context({'input': user_input}, {'output': response})

        print(f"Chatbot: {response}")

if __name__ == "__main__":
    main()
```
